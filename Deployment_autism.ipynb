```python
import pandas as pd
import pycaret
dataset=pd.read_csv('/content/drive/MyDrive/autism project/dataset (1).csv')
```

```python
train=dataset.sample(frac=0.8,random_state=10)
test=dataset.drop(train.index)
```

```python
train.shape,test.shape
```

```python
train
```

```python
#basic preprocessing 
train=train.drop('age_desc',axis=1)
train=train.drop('result',axis=1)
train=train.replace(to_replace=['YES}'],value='YES')
train['relation']=train['relation'].str.replace('?','Self')
train['ethnicity']=train['ethnicity'].str.replace('?','White-European')
train['age']=train['age'].str.replace('?','21')
```

```python
#all preprocessing 
from pycaret.classification import *
setting_config= setup(data=train,target='Class_ASD',session_id=123)
```

```python
#model selection
compare_models()
```

```python
#model creation
ada=create_model('ada')
```

```python
#hyperparameter tuning 
tuned_model=tune_model(ada)
```

```python
#finalised model 
final_model=finalize_model(tuned_model)
```

```python
#predictions
predict_model(final_model,test)
```

```python
save_model(final_model,"Best_model")
```

```python
final=load_model("Best_model")
```
